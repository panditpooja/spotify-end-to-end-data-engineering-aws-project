The above info is for one item. Now lets write a generic user defined functions that can help extract data for all the albums

```python
ID=[]
Album_Name=[]
Release_Date=[]
Total_Tracks=[]
External_URL=[]
```

```python
for item in data['items']:
  ID.append(item['track']['album']['id'])
  Album_Name.append(item['track']['album']['name'])
  Release_Date.append(item['track']['album']['release_date'])
  Total_Tracks.append(item['track']['album']['total_tracks'])
  External_URL.append(item['track']['album']['external_urls']['spotify'])
```

```python
import pandas as pd

df_album=pd.DataFrame({
    'ID':ID,
    'Album_Name':Album_Name,
    'Release_Date':Release_Date,
    'Total_Tracks':Total_Tracks,
    'External_URL':External_URL
    })
```

```python
df_album.head()
```

The above way of creating df_album is easy to interpret but in not efficient for very large datasets as it needs intermediate list creation resulting in more memory usage. The better option is using dictionary and then creating dataframes from them. Below is the code.

```python
album_list=[]
for item in data['items']:
  ID=item['track']['album']['id']
  Album_Name= item['track']['album']['name']
  Release_Date=item['track']['album']['release_date']
  Total_Tracks=item['track']['album']['total_tracks']
  External_URL=item['track']['album']['external_urls']['spotify']
  album_list.append({'ID':ID,'Album_Name':Album_Name,'Release_Date':Release_Date,'Total_Tracks':Total_Tracks, 'External_URL':External_URL})
```

```python
album_list
```

Lets start extracting data pertaining to Artist info

```python
data['items'][0]['track']['album']['artists']
```

```python
data['items'][3]['track']['album']['artists']
```

```python
artist_list=[]
for item in data['items']:
  artist=item['track']['album']['artists']
  for each_artist in artist:
    artist_id=each_artist['id']
    artist_name=each_artist['name']
    external_url=each_artist['href']
    artist_list.append({'Artist_ID':artist_id,'Artist_Name':artist_name,'External_URL':external_url})
artist_list
```

Lets start extracting data pertaining to Songs info

```python
data['items'][0]['track']
```

```python
songs_list=[]
for item in data['items']:
  song_id=item['track']['id']
  song_name=item['track']['name']
  song_duration_ms=item['track']['duration_ms']
  song_url=item['track']['external_urls']['spotify']
  song_popularity=item['track']['popularity']
  song_added=item['added_at']
  album_id=item['track']['album']['id']
  artist_id=item['track']['album']['artists'][0]['id']
  songs_list.append({'Song_ID':song_id,'Name':song_name,'Duration_ms':song_duration_ms,'URL':song_url,'Popularity':song_popularity,'Added_At':song_added,'Album_ID':album_id,'Artist_ID':artist_id})
songs_list
```

Solution:
The warning you're encountering suggests that you're trying to modify a slice of the DataFrame directly, which can lead to unexpected behavior. You can fix this by using .loc[row_indexer, col_indexer] to explicitly assign the values. Here's how you can do it:

album_df.loc[:, 'Release_Date'] = pd.to_datetime(album_df['Release_Date'])

This explicitly tells pandas to modify the original DataFrame rather than a slice of it, avoiding the warning.
